###### Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)

```python
avg_this_wk=merged_df["CodeKata Score"].mean()
print("Average of this week geek coins is: ",avg_this_wk)
avg_pvs_wk=merged_df["Previous Geekions"].mean()
print("Average of previous week geek coins is: ",avg_pvs_wk)
rise_in_avg=avg_this_wk-avg_pvs_wk
print("Rise in average this week is: ",rise_in_avg)
fig=px.scatter(merged_df,x="CodeKata Score",y="Previous Geekions",title="Codekata Score vs Previous Geekions")
fig.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
topper_name=[]
x1=merged_df[merged_df["Department"]=="Computer Science and Engineering"]
csmax=x1["CodeKata Score"].max()
topper_name.append(x1.Name[x1["CodeKata Score"]==x1["CodeKata Score"].max()].values[0])
x2=merged_df[merged_df["Department"]=="Electronics and Communication Engineering"]
ecmax=x2["CodeKata Score"].max()
topper_name.append(x2.Name[x2["CodeKata Score"]==x2["CodeKata Score"].max()].values[0])
x3=merged_df[merged_df["Department"]=="Electronics and Electrical Engineering"]
eemax=x3["CodeKata Score"].max()
topper_name.append(x3.Name[x3["CodeKata Score"]==x3["CodeKata Score"].max()].values[0])
```

```python
print("The department wise toppers are:")
print(topper_name)
```

```python
l1=["Computer Science","Electronics and Communication Engineering","Electronics and Electrical Engineering"]
l2=[csmax,ecmax,eemax]
print("Department wise toppers names")
fig=px.bar(x=l1,y=l2, color=topper_name,title="Department wise Topper")
fig.update_layout(
    title="Department wise toppers "
)
fig.show()
```
